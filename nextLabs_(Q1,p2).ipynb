```python
import pandas as pd
import re
import nltk
```

```python
df = pd.read_csv("/content/chrome_reviews.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Text'] = df['Text'].fillna("Good", inplace=False)
```

```python
df['Text'][5163]
```

```python
df2 = df.copy()
```

```python
df2['Text'].head(10)
```

```python
def de_emojify(text):
    regex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F" 
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
                           "]+", flags = re.UNICODE)
    return regex_pattern.sub(r'', text)
```

```python
de_emojify(df2['Text'][8])
```

```python
for i in range(0,7204):
  df2['Text'][i] = de_emojify(df['Text'][i])
```

```python
df2.head(10)
```

```python
df3 = df2.iloc[:,2:4]
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

PS = PorterStemmer()

stop_words = stopwords.words("english")
stop_words.remove("not")
```

```python
for i in range(0,7204):
  r = re.sub("[^a-zA-Z]",' ', df2['Text'][i])
  r = r.lower()
  r = r.split()
  r = [PS.stem(j) for j in r if i not in set(stop_words)]
  r = ' '.join(r)

  df3['Text'][i] = r
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
df4 = df3.iloc[0:1]
df4
```

```python
def sentiment_analyse(sent_text):
  score = SentimentIntensityAnalyzer().polarity_scores(sent_text)
  print(score)
```

```python
df3['Text'][5871]
```

```python
sentiment_analyse(df3['Text'][5871])
```

```python
df3['Text'][6311]
```

```python
for i in range(0,7204):
  score = SentimentIntensityAnalyzer().polarity_scores(df3['Text'][i])
  if score['pos']>=0.50 and score['neu']<0.30:
    df4.loc[df3.index[i]] = df3.iloc[i]
```

```python
df4 = df4.drop(df4.index[0], axis=0)
```

```python
df5 = df4[df4['Star']<=2]
```
