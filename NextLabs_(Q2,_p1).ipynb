```python
from happytransformer import HappyTextToText
```

```python
happy = HappyTextToText("T5","prithivida/grammar_error_correcter_v1")
```

```python
from happytransformer import TTSettings
```

```python
settings = TTSettings(do_sample=True, top_k=50, temperature=0.7, min_length=1, max_length=50)
```

```python
df2 = df.copy()
```

```python
for i in range(0,30000):
  df2['text'] = happy.generate_text(df['text'][i], args=settings)
```
